# 1. 데이터 마이닝이란?
<li>방대한 양의 데이터 속에서 숨겨진 규칙，패턴 등을 찾아내어 예측
하거나 의사결정에 활용히는 것을 목적</li>
<li>컴퓨터가 스스로 학습하는 알고리즘인 머신러닝 (Machine Learning)을 구현하기 위한 바탕</li>

# 2. 데이터 마이닝 종류 ***

## 1) 지도학습(문제 + 정답)
### <li>정답이 있는 데이터를 활용해 분석 모댈을 학습시키는 것</li>
### <li> 고양이 사진 50장과 강아지 사진 50장을 입력 데이터로 줄 때 각 사진마다 고양이 사진인지， 강아지 사진인지에 대한 정보(일종의 정답지)가 필요</li>

## 2) 비지도학습(문제만)
### <li>지도학습과는 달리 정답을 알려주지 않고 학습하는 것</li>
### <li>정답 레이블이 없는 강아지 사진과 고양이 사진 100장을 입력 데이터로 줄 때， 컴퓨터는 정확히 어떤 동물이라고 정의할 수는 없지만 동물들의 특징이 비슷한 동물끼리 군집화를 실시</li>
