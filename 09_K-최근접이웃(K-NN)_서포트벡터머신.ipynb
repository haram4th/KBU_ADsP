# 1. K-최근접이웃(K-NN, K-Nearest Neighbor)
## 1) K-최근접이웃 알고리즘
* 지도학습인 분류 분석에 속하지만 실은 군집의 특성도 가지고 있어 책에 따라서는 Semi(준)-지도학습으로 분류
* k-NN은 정답 라벨이 없는 새로운 데이터를 입력 받았을 때 그 데이터로부터 가장 가까이에 있는 데이터의 정답 라벨을 확인하여 새로운 데이터의 정답 라벨을 결정

# 2. 서포트벡터머신(SVM : Support Vector Machine)
* 지도학습에 주로 이용되며 특히 분류 성능이 뛰어나 분류 분석에 자주 사용
* 고차원 또는 무한 차원의 공간에서 초평면의 집합을 찾아 여백을 최대화 하여 분류한다.
